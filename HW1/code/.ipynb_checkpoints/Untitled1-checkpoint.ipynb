```python
import matplotlib.pyplot as plt

# create a list of 27 numbers
my_list = [94.6, 75.1, 88.6, 94.7, 75, 89, 94.7, 75, 88.6, 95, 76.1, 90.5, 95, 76, 90.5, 95, 76, 90.5, 95, 76.2, 90.6, 95, 76, 90.6, 95, 76, 90.6]

# create a new figure
fig = plt.figure()

# create a subplot
ax = fig.add_subplot(111)

# plot the list as bars
ax.bar(range(len(my_list)), my_list, width=0.8, color='k')

# set the x-axis label
ax.set_xlabel('Index')

# set the y-axis label
ax.set_ylabel('Accuracy')

# set the title of the plot
ax.set_title('Accuracy Plot')

# display the plot
plt.show()
```
